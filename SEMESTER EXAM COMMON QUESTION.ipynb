```python
import pandas as pd
import numpy as np

!pip install pandas numpy

try:
    df = pd.read_csv('/content/AQI_Data (3).csv')
except FileNotFoundError:
    print("Error: 'air_quality_data.csv' not found. Please ensure the file exists and the path is correct.")
    df = pd.DataFrame(columns=['City', 'AQI', 'PM2.5', 'PM10'])
    exit()


print("First 5 rows:")
print(df.head())


print("\nLast 6 rows:")
print(df.tail(6))



print("\nSummary statistics for numeric columns:")
print(df.describe())



print("\nMean AQI, PM2.5, and PM10 values for each city:")
if 'City' in df.columns and 'AQI' in df.columns and 'PM2.5' in df.columns and 'PM10' in df.columns:
  for city in df['City'].unique():
    city_data = df[df['City'] == city]
    mean_aqi = np.mean(city_data['AQI'])
    mean_pm25 = np.mean(city_data['PM2.5'])
    mean_pm10 = np.mean(city_data['PM10'])
    print(f"City: {city}")
    print(f"  Mean AQI: {mean_aqi:.2f}")
    print(f"  Mean PM2.5: {mean_pm25:.2f}")
    print(f"  Mean PM10: {mean_pm10:.2f}")
else:
    print("Error: One or more required columns ('City', 'AQI', 'PM2.5', 'PM10') not found in the DataFrame.")
```
